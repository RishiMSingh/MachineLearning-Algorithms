# Data Understanding

## Information from the readme file has helped the data understanding of this project:

- instant: record index
- dteday : date
- season : season (1:springer, 2:summer, 3:fall, 4:winter)
- yr : year (0: 2011, 1:2012)
- mnth : month ( 1 to 12)
- hr : hour (0 to 23)
- holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)

+ weathersit :

	- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : Normalized temperature in Celsius. The values are divided to 41 (max)
- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
- hum: Normalized humidity. The values are divided to 100 (max)
- windspeed: Normalized wind speed. The values are divided to 67 (max)
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered

```python
class RidgeRegression:
    def __init__(self, learning_rate=0.01, alpha=0.1, n_iters=1000):
        self.lr = learning_rate
        self.alpha = alpha
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # Initialize parameters
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Gradient descent
        for _ in range(self.n_iters):
            # Calculate predictions
            y_predicted = np.dot(X, self.weights) + self.bias
            
            # Compute gradients
            dw = (1 / n_samples) * (2 * np.dot(X.T, (y_predicted - y)) + 2 * self.alpha * self.weights)
            db = (1 / n_samples) * 2 * np.sum(y_predicted - y)
            
            # Update parameters
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
# Instantiate the Ridge Regression model
ridge_reg = RidgeRegression(learning_rate=0.001, alpha=1.0, n_iters=1000)

# Assuming 'X_train' is the training data and 'y_train' is the target values
ridge_reg.fit(X_train, y_train)

# To predict values with the model
y_pred = ridge_reg.predict(X_test)

# Evaluate the model
mse = np.mean((y_test - y_pred)**2)
print("MSE:", mse)
```

```python
X_train_array = X_train.to_numpy()
y_train_array = y_train.to_numpy()

class RidgeRegressionMBGD:
    def __init__(self, learning_rate=0.01, alpha=0.1, n_iters=1000, batch_size=32):
        self.lr = learning_rate
        self.alpha = alpha
        self.n_iters = n_iters
        self.batch_size = batch_size
        self.weights = None
        self.bias = None
        self.mse_history = []

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            indices = np.arange(n_samples)
            np.random.shuffle(indices)
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            for i in range(0, n_samples, self.batch_size):
                end = i + self.batch_size
                X_batch = X_shuffled[i:end]
                y_batch = y_shuffled[i:end]

                y_predicted = np.dot(X_batch, self.weights) + self.bias
                error = y_predicted - y_batch
                dw = (1 / n_samples) * (2 * np.dot(X_batch.T, error) + 2 * self.alpha * self.weights)
                db = (1 / n_samples) * 2 * np.sum(error)

                self.weights -= self.lr * dw
                self.bias -= self.lr * db

            y_predicted_full = self.predict(X)
            mse = self.mean_squared_error(y, y_predicted_full)
            self.mse_history.append(mse)

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

    def mean_squared_error(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)
```

```python
# Make sure X_train and y_train are numpy arrays
ridge_reg_mbgd = RidgeRegressionMBGD(learning_rate=0.001, alpha=1.0, n_iters=1000, batch_size=32)
ridge_reg_mbgd.fit(X_train_array, y_train_array)

#Plotting the MSE history
plt.plot(range(1, ridge_reg_mbgd.n_iters + 1), ridge_reg_mbgd.mse_history)
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('MSE over Epochs for Ridge Regression with Mini-Batch Gradient Descent')
plt.show()
```
